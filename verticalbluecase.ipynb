### Potential Market Size & CAGR Forecast

```python
df3 = pd.read_csv("files/trend_lang_app.csv",index_col="YEAR",
                 infer_datetime_format=True,
                 parse_dates=True)

df3 = df3.drop('Unnamed: 2', axis=1)

df3 = df3.reset_index()

df3.columns = ["ds", "y"]

model = Prophet()

model.fit(df3)

future_trends = model.make_future_dataframe(periods=10, freq="Y")

forecast_trends = model.predict(future_trends)

model.plot(forecast_trends)
```

```python
cagrdf = forecast_trends[['ds','yhat']]

cagr_eq = (cagrdf['yhat'].iloc[-1]/8.21)**(1/9) - 1

print(cagrdf['yhat'][0])
print(cagrdf['yhat'].iloc[-1])
print(cagr_eq)
print(cagrdf)
```
